<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
# @@@ 5
# Expenditure per Student
df["expenditure_per_student"] = df["TOTAL_EXPENDITURE"] / df["ENROLL_ALL"]
# Average Math Score Change Over Time
df = df.sort_values(["STATE", "YEAR"])
df["math_score_change"] = df.groupby("STATE")["AVG_MATH_4_SCORE"].diff()
# Relative Performance to National Average
national_mean = df.groupby("YEAR")["AVG_MATH_4_SCORE"].transform("mean")
df["math_score_vs_national"] = df["AVG_MATH_4_SCORE"] - national_mean



```

Feature engineering justification: Total spending alone can be misleading because states differ in enrollment size. To account for this, I normalized total expenditure by the number of students. To capture performance trends rather than raw scores, I computed the year-to-year change in average math scores by state. I also created a feature measuring how each state’s performance compares to the national mean in the same year.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
# @@@ 8

X = df[['ENROLL_4','expenditure_per_student','AVG_READING_4_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_4_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
# @@@ 9 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# @@@ 10
from sklearn.linear_model import LinearRegression
```

```python
# @@@ 11
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
# @@@ 12
print(model.score(X_test, y_test))
print(np.mean(model.predict(X_test)-y_test))
print(np.mean(np.abs(model.predict(X_test)-y_test)))
print(np.mean((model.predict(X_test)-y_test)**2)**0.5)
```
